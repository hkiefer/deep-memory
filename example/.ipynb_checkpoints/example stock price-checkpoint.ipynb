```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from deep_memory import *
```

```python
#load stock price as dataframe with yfinance module in deep_memory
trj = loaddata('FB', interval = 'daily')
trj['Close'].plot()
```

```python
#calculating xva_frame and correlation functions
xf=xframe(trj['Close'],trj['Close'],fix_time=True)
xvaf=compute_va(xf,correct_jumps=True)

x = xvaf["x"]
v = xvaf["v"]
a = xvaf['a']

corrv=correlation(v)
corrva = correlation(v,a)
corra = correlation(a)
```

```python
#deep memory kernel extraction with Multi-Layer-Perceptron

trunc = 1000
X = corrv[:trunc]

#y = np.gradient(x)
y = corrva[:trunc]

X = np.array([X]).T
y = np.array([y]).T
print(X.shape, y.shape)

plt.plot(X)
plt.xscale('log')
plt.title('input')
plt.show()

plt.plot(y)
plt.xscale('log')
plt.title('output to predict')
plt.show()
```

```python
#fitting MLP (with given initial value of memory kernel)

model = MultilayerPerceptron(n_hidden=3,
    n_epochs=200,
    learning_rate=0.0001, dt = 1, fe = None, optimizer = 'Rprop')

loss, model_fit_deep = model.fit(X, y, accept_bias = True, initial = None, early_stop = 1e-6)
```

```python
dt = 0.01
kernel_index = np.arange(0,trunc,0.01)
kernel_deep = model.compute_kernel(kernel_index)


plt.plot(kernel_index,kernel_deep,label = 'deep memory')

plt.xscale('log')
plt.ylabel(r'$\Gamma$(t)', fontsize = 'x-large')
plt.xlabel('t', fontsize = 'x-large')
plt.xticks(fontsize = 'x-large')
plt.yticks(fontsize = 'x-large')
plt.legend()
plt.show()
```

```python
#learning curve
epochs = np.arange(1,len(loss)+1)
plt.plot(epochs, loss)
plt.ylabel('MSE')
plt.xlabel('epoch')
plt.yscale('log')
```
